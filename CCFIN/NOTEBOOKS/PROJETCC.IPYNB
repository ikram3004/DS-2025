```python
from ucimlrepo import fetch_ucirepo
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger le dataset
student_performance = fetch_ucirepo(id=320)

# Extraire les données
X = student_performance.data.features
y = student_performance.data.targets

# Créer le dataframe complet
df = pd.concat([X, y], axis=1)

# Distribution des notes finales (G3)
plt.figure(figsize=(10, 6))
sns.histplot(y['G3'], bins=20, kde=True)
plt.title('Distribution des Notes Finales (G3)')
plt.xlabel('Note (0-20)')
plt.ylabel('Fréquence')
plt.show()

# Corrélation entre les variables
plt.figure(figsize=(12, 10))
correlation_matrix = df.corr(numeric_only=True)  # Ajout de numeric_only=True
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm')
plt.title('Matrice de Corrélation')
plt.show()
```
